# Setup check

Confirms the environment works and PyTorch can reach the Apple Silicon GPU.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch

device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")
print(f"torch {torch.__version__} on {device}")
```

```python
# Tiny end-to-end sanity check: fit y = 3x + 2 on the GPU.
x = torch.linspace(-1, 1, 200, device=device).unsqueeze(1)
y = 3 * x + 2 + 0.1 * torch.randn_like(x)

model = torch.nn.Linear(1, 1).to(device)
opt = torch.optim.Adam(model.parameters(), lr=0.1)

for step in range(200):
    loss = torch.nn.functional.mse_loss(model(x), y)
    opt.zero_grad()
    loss.backward()
    opt.step()

w, b = model.weight.item(), model.bias.item()
print(f"learned y = {w:.2f}x + {b:.2f}  (loss {loss.item():.4f})")
```

```python
plt.scatter(x.cpu(), y.cpu(), s=8, alpha=0.5, label="data")
plt.plot(x.cpu(), model(x).detach().cpu(), color="crimson", label="fit")
plt.legend()
plt.show()
```
